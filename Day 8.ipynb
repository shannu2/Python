```python
n=int(input("enter n value"))
array = []  
for i in range(n):           
    element=(int(input("enter an element"))) 
    array.append(element)
for i in range(n-1):
    for j in range(n-i-1):
        if(array[j]>array[j+1]):
            temp=array[j]
            array[j]=array[j+1]
            array[j+1]=temp
for i in range(n):
     print(array[i], end = " ") 
```

```python
n=int(input("enter n value"))
array = []  
for i in range(n):           
    element=(int(input("enter an element"))) 
    array.append(element)
for i in range(n):  
        print(array[i], end = " ") 
key=int(input("enter key value"))
low=0
high=n-1
mid=(low+high)//2
while(low<=high):
    if(key<array[mid]):
        high=mid-1
    elif(key==array[mid]):
        print("key is found at:",mid)
        break
    elif(key>array[mid]):
        low=mid+1
        mid=(low+high)//2
```
